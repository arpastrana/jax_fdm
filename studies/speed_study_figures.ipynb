```python
import numpy as np
import matplotlib.pyplot as plt

import pickle
```

```python
arches_info = pickle.load(open('arches_info.pkl', 'rb'))
tower_info = pickle.load(open('tower_info.pkl', 'rb'))
saddle_info = pickle.load(open('saddle_info.pkl', 'rb'))
```

```python
arches_info[0].keys()
```

```python
info_list = arches_info

num_segments = []

sparse_lower = []
sparse_upper = []
sparse_median = []

dense_lower = []
dense_upper = []
dense_median = []

for info in info_list:
    sorted_sparse = sorted(info['sparse_times'])
    sparse_lower.append(sorted_sparse[0])
    sparse_upper.append(sorted_sparse[-1])
    sparse_median.append(sorted_sparse[2])

    sorted_dense = sorted(info['no_sparse_times'])
    dense_lower.append(sorted_dense[0])
    dense_upper.append(sorted_dense[-1])
    dense_median.append(sorted_dense[2])
    
    num_segments.append(info['num_segments'])

sparse_lower = np.array(sparse_lower)
sparse_upper = np.array(sparse_upper)
sparse_median = np.array(sparse_median)

dense_lower = np.array(dense_lower)
dense_upper = np.array(dense_upper)
dense_median = np.array(dense_median)

plt.title("Arches Study")
plt.yscale('log'); plt.xscale('log')
plt.xlabel('Number of Segments')
plt.ylabel('seconds (s)')

plt.errorbar(x=num_segments, y=sparse_median,
             yerr=np.stack((sparse_median-sparse_lower, sparse_upper-sparse_median)),
             capsize=3.0, label='Sparse')

plt.errorbar(x=num_segments, y=dense_median,
             yerr=np.stack((dense_median-dense_lower, dense_upper-dense_median)),
             capsize=3.0, label='Dense')

plt.legend();
```
